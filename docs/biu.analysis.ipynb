## RNA-Seq

### Normalization

Some normalization schemes are implemented here:
 * TMM
 * CPM (voom)
 
Often these are used together.

```python
D1 = pd.DataFrame(np.abs(np.random.normal(20, 10, (10, 500))), index=['A%d' % i for i in range(10)])
D2 = pd.DataFrame(np.abs(np.random.normal(40, 10, (10, 500))), index=['B%d' % i for i in range(10)])
D  = pd.concat([D1, D2])
biu.analysis.rnaseq.normalize('voom', biu.analysis.rnaseq.normalize('TMM', D))
```
