# Multiple Hypothesis Testing Correction procedures

We implement a few hypothesis correction procedures in BIU:

 * FWER correction
  * bonferroni correction
 * FDR correction
  * Benjamini-Hochberg
  * Benjamini-Hochberg-Yekutieli
  
These procedures are available in the `biu.stats.correction.correct` function.

```python
import biu
import numpy as np

pValues = np.random.uniform(high=0.03, low=0.0, size=100)
pValues = sorted(pValues)
```

## FWER correction

### Bonferroni correction

```python
qValuesFWER = biu.stats.correction.correct(pValues, 'FWER')

fig, axes = biu.utils.figureUtils.subplots(figsize=(9,4.5), ncols=1, nrows=1, dpi=500)

axes[0].plot(pValues, qValuesFWER, '*')
axes[0].set_xlabel('p-values')
axes[0].set_ylabel('FWER corrected q-values')
None
```

## FDR correction

### Benjamini Hochberg

```python
qValuesBH = biu.stats.correction.correct(pValues, 'fdr_bh')

fig, axes = biu.utils.figureUtils.subplots(figsize=(9,4.5), ncols=1, nrows=1, dpi=500)

axes[0].plot(pValues, qValuesBH, '*')

axes[0].set_xlabel('p-values')
axes[0].set_ylabel('FDR BH corrected q-values')
None
```

### Benjamini Hochberg Yekutieli

With the BHY procedure, you can control for dependencies between the tests.
By default, the implemented BHY procedure corrects for arbitrary dependence, but you can specify a different `c(m)` with the `cm=` option.

```python
qValuesBHY = biu.stats.correction.correct(pValues, 'fdr_bhy')

fig, axes = biu.utils.figureUtils.subplots(figsize=(9,4.5), ncols=1, nrows=1, dpi=500)

axes[0].plot(pValues, qValuesBHY, '*')

axes[0].set_xlabel('p-values')
axes[0].set_ylabel('FDR BHY corrected q-values')

None
```

Setting `cm=1` results in the BH procedure

```python
qValuesBHY = biu.stats.correction.correct(pValues, 'fdr_bhy', cm=1)

fig, axes = biu.utils.figureUtils.subplots(figsize=(9,4.5), ncols=1, nrows=1, dpi=500)

axes[0].plot(pValues, qValuesBHY, '*')

axes[0].set_xlabel('p-values')
axes[0].set_ylabel('FDR BHY corrected q-values, cm=1')

None
```
