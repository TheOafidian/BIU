# Figure utilities in BIU

```python
import biu
```

## Wrapper for matplotlib's subplot function

Matplotlibs subplot function returns either an array of axes or a single axis. This function makes it behave in a normal way.

```python
fig, axes = biu.utils.figure.subplots(figsize=(3,4), nrows=1, ncols=2)
axes[0].plot([0,10,20], [10, 0, 20])
axes[1].plot([10, 0, 20], [0,10,20])
fig.show()
```

```python
fig, axes = biu.utils.figure.subplots(figsize=(3,4), nrows=1, ncols=1)
axes[0].plot([0,10,20], [10, 0, 20])
fig.show()
```
