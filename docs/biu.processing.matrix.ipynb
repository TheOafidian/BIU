# Matrix operations

```python
import biu
import numpy as np
import matplotlib.pylab as plt
```

## Reorder matrix rows

It can be useful to re-order matrix rows based on a clustering. For this, we can use the `biu.processing.matrix.order` function.

```python
data1 = np.concatenate([ np.random.uniform(low=10, high=25, size=(30,30)), 
                         np.random.uniform(low=30, high=100, size=(30,30)),
                         np.random.uniform(low=-20, high=15, size=(30,30))])

# Shuffle the data
randOrder = list(range(data1.shape[0]))
np.random.shuffle(randOrder)
data2 = data1[randOrder,:]

# Reorder based on clustering
reOrder = biu.processing.matrix.order(M=data2, distance='euclidean', method='single')
data3 = data2[reOrder,:]

fig, axes = biu.utils.figure.subplots(figsize=(30,20), nrows=1, ncols=3)

axes[0].imshow(data1)
axes[0].set_title("Dataset")
axes[1].imshow(data2)
axes[1].set_title("Dataset -> shuffled")
axes[2].imshow(data3)
axes[2].set_title("Dataset -> shuffled -> reordered")
plt.show()
```
