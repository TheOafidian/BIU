# Handling FASTA files


```python
import biu
```

```python
fasta = biu.formats.Fasta('example_files/example.fasta')

print(fasta)
```

```python
for seqid in fasta:
    print('>%s\n%s\n' % (fasta[seqid].fullName, fasta[seqid]))
```

```python
print(fasta["HSGLTH1"])
```

```python
fasta["newSeq"] = "GGCAGATTCCCCCTAGAC"

fasta["newSeq"]
```

```python
fastaProts = biu.formats.Fasta( [ fasta[s].translate() for s in fasta ] )

print(fastaProts)
```

```python
fastaDna = biu.formats.Fasta( [ fastaProts[s].reverseTranslate() for s in fastaProts ] )

print(fastaDna)
```
