# Accessing the COSMIC database

```python
import biu as biu

where = '/exports/molepi/tgehrmann/data/'
biu.config.settings.setWhere(where)

cosmic = biu.db.Cosmic("t.gehrmann@lumc.nl", "Cosmic_password1", version="grch37_85")
print(cosmic)
```

## Querying COSMIC
You can query cosmic the same way you can query any other VCF structure, using the query function:

```python
for r in cosmic.query(1, 1, 69270):
    print(r)
```

### Only coding

```python
for r in cosmic.query(1, 1, 69270, noncoding=False):
    print(r)
```

### Only noncoding

```python
for r in cosmic.query(1, 1, 69270, coding=False):
    print(r)
```
