# Interpolation

```python
import biu
import numpy as np
```

# Linear interpolation
Linearly interpolate points given a curve

```python
x = np.arange(0, 3*np.pi, 0.1)
y = np.sin(x)

fig, ax = biu.utils.figure.subplots(nrows=1, ncols=2, figsize=(10,5))

ax[0].scatter(x,y)

ax[1].scatter(np.arange(0, 3*np.pi, 0.01), 
              biu.math.interpolation.linearInterpolation(x, y, np.arange(0, 3*np.pi, 0.01)))
```
