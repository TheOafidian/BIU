# Regression

```python
import biu
import numpy as np
import matplotlib.pylab as plt
import scipy as sp
```

## LOWESS regression

```python
# Generate data
group1 = np.random.multivariate_normal(mean=[0,0], cov=[[1,0],[0,1]], size=100)
group2 = np.random.multivariate_normal(mean=[4,-4], cov=[[1,0],[0,1]], size=100)
data = np.concatenate([group1, group2])

fig, ax = biu.utils.figure.subplots(nrows=1, ncols=2, figsize=(10,5))

# Fit the curve for all points
yest = biu.stats.regression.lowess(data[:,0], data[:,1])

ax[0].scatter(data[:,0], data[:,1])
ax[0].scatter(data[:,0], yest, c='r')

# Get a linearly interpolated function
lowessFunc = sp.interpolate.interp1d(data[:,0], yest)

ax[1].scatter(data[:,0], data[:,1])
ax[1].plot(sorted(data[:,0]), lowessFunc(sorted(data[:,0])), c='r')
```
